```python
import geopandas as gpd
import folium

# Load the GeoJSON
gdf = gpd.read_file(str(sources[0].resolve()))

# Ensure CRS is WGS84 (required for folium)
gdf = gdf.to_crs(epsg=4326)

# Create a base map centered on the dataset
center = [
    gdf.geometry.centroid.y.mean(),
    gdf.geometry.centroid.x.mean()
]

m = folium.Map(
    location=center, 
    zoom_start=13, 
    tiles="CartoDB dark_matter"#"CartoDB positron"
)


color_map = {
    "CONTROLLED ACCESS HIGHWAY": "#d73027",
    "ARTERIAL STREET": "#fc8d59",
    "COLLECTOR STREET": "#fee08b",
    "PROPOSED FUTURE ARTERIAL STREET": "#f46d43",
    "PROPOSED FUTURE COLLECTOR STREET": "#fdae61",
    "PROPOSED BIKE PATH": "#2b83ba",
}

def style_function(feature):
    val = feature["properties"].get("OP_TRANSP")
    
    dash = "5,5" if "PROPOSED" in str(val) else None
    
    color = color_map.get(val, "#999999")
    
    return {
        "fillColor": color,
        "color": color,
        "weight": 4,
        "dashArray": dash,
        "fillOpacity": 0.6
    }

# Add GeoJSON with automatic tooltip fields
columns = [c for c in gdf.columns if c != "geometry"]

# Add GeoJSON layer
# Add styled layer
folium.GeoJson(
    gdf,
    style_function=style_function,
    tooltip=folium.GeoJsonTooltip(fields=columns, aliases=columns) # (fields=["OP_TRANSP"])
).add_to(m)

# Save map
m.save("map.html")

print("Map saved to map.html")
```
